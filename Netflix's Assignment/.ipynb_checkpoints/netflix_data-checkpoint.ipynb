```python
import pandas as pd
from sqlalchemy import create_engine, types
csv_file = "Netflix_titles.csv"
df = pd.read_csv(csv_file, encoding="utf-8")

connection_string = "mssql+pyodbc://DESKTOP-6HRT637/master?driver=ODBC+Driver+17+for+SQL+Server"
engine = create_engine(connection_string, fast_executemany=True)

sql_types = {
    "show_id": types.NVARCHAR(10),
    "type": types.VARCHAR(10),
    "title": types.NVARCHAR(200),
    "director": types.NVARCHAR(250),
    "cast": types.NVARCHAR(1000),    
    "country": types.NVARCHAR(150),
    "date_added": types.VARCHAR(30),
    "release_year": types.INTEGER(),
    "rating": types.VARCHAR(10),
    "duration": types.VARCHAR(20),
    "listed_in": types.NVARCHAR(100),
    "description": types.NVARCHAR(1000)
}
try:
    df.to_sql(
        name="Netflix_raw",
        con=engine,
        schema="dbo",
        if_exists="replace", 
        index=False,
        chunksize=1000,
        dtype=sql_types     
    )
    print("Success! Data with Korean characters has been uploaded.")
except Exception as e:
    print(f"An error occurred: {e}")
```
