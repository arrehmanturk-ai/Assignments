## Looping Structure

```python
for i in range(1, 11):
    print(i)
```

```python
i = 20 
while i >= 1:
    print(i)
    i -= 1
```

```python
i = 1
while i <= 10:
    if (i%2 !=0):
        i += 1
        continue
    print(i)
    i += 1
```

```python
n = int(input("Enter a Number :"))
i = 1
while i <= n:
    print(i)
    i += 1
```

```python
n = int(input("Enter a Number:"))

i = 1
while i <= n:
    if (i%2 == 0):
        i += 1
        continue
    print(i)
    i += 1
```

```python
i = 1
while i <= 5 :
        print("Happy Birthday!")
        i += 1
```

```python
n = int(input("Enter a Number :"))
squaring = []
for i in range(1, n+1):
    squaring.append(i**2)

print("Squares of Numbers =", squaring)
```

```python
n = int(input("Enter a Number :"))
print(f"The Table of Number :", n)

for i in range(1, 11):
    print(f"{n} * {i} = {n*i}")
   # print(n*i)
```

```python
a = 3
d = 4
n = 8

for i in range(n):
    term = a + i * d
    print(term, end=' ')
```

```python
a = 2
r = 3
n = 6

for i in range(n):
    term = a * (r ** i)
    print(term, end=' ')
```

```python
n = int(input("Enter a Number :"))
sum = 1

for i in range(1, n+1):
    sum += i
print("The sum of the Numbers =", sum)
```

```python
n = int(input("Enter a Positive Integer :"))
total = 0.00

for i in range(1, n+1):
    total += 1/i

print(f"The sum of reciprocals from 1 to {n} is : {total:.2f}")
```

```python
total = 0
for i in range(5):
    num = int(input("Enter a Number :"))
    total += num

print("The final running total is :", total)
```

```python
fact = 1
for i in range(5):
    num = int(input("Enter a Number :"))
    fact *= num

print("The final running factorial is :", fact)
```

```python
base = float(input("Enter the base number: "))
exponent = int(input("Enter the exponent: "))

result = 1.0
abs_exponent = abs(exponent)

for _ in range(abs_exponent):
    result *= base

if exponent < 0:
    result = 1 / result

print(f"{base} power {exponent} = {result}")
```
