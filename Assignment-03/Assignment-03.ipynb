# Strings Assignments

```python
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for char in text:
    if char in vowels:
        count += 1
print("Number of vowels in the string:", count)
```

```python
text = input("Enter a string:")
uppercase = 0
lowercase = 0
digits = 0
whitespace = 0

for char in text:
    if char.isupper():
        uppercase += 1
    elif char.islower():
        lowercase += 1
    elif char.isdigit():
        digits += 1
    elif char.isspace():
        whitespace += 1
print("Number of uppercase letters:", uppercase)
print("Number of lowercase letters:", lowercase)
print("Number of digits:", digits)
print("Number of whitespace characters:", whitespace)
```

```python
original_string = input("Enter a string: ")

if len(original_string) < 2:
    print("String must have at least 2 characters to exchange first and last characters.")
else:
    new_string = original_string[-1] + original_string[1:-1] + original_string[0]

print(f"Original string: {original_string}")
print(f"New string: {new_string}")
```

```python
original_string = input("Enter a string: ")
new_string = original_string[::-1]

print(f"Original string: {original_string}")
print(f"New string: {new_string}")
```

```python
original_string = input("Enter a string: ")
new_string = original_string[1:] + original_string[0]

print(f"Original string: {original_string}")
print(f"New string: {new_string}")
```

```python
full_name = input("Enter your First, Middle and Last name :")
initials = ""
i = 0

while i < len(full_name):
    initials += full_name[i].upper() + ". "
    while i < len(full_name) and full_name[i] != ' ':
        i += 1
    while i < len(full_name) and full_name[i] == ' ':
        i += 1
        
print("Initials:", initials)
```

```python
user = input("Enter the word: ")
i = 0
j = len(user) - 1

while i < j:
    if user[i] != user[j]:
        print("It's not a palindrome")
        break
    else:
        i += 1
        j -= 1
else:
    print(f"{user} a palindrome")
```

```python
user = input("Enter your password:".title())
passwords = ""
uppercase = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
lowercase = "abcdefghijklnopqrstuvwxyz"
digits = "0123456789"

has_uppercase = False
has_lowercase = False
has_digits = False

if len(user) >= 8 :
    for char in user:
        if char in uppercase :
           has_uppercase = True
        elif char in lowercase :
            has_lowercase = True
        elif char in digits:
            has_digits = True

    if has_uppercase and has_lowercase and has_digits :
        print("the password is valid".title())
    else:
        if not has_uppercase:
            print("atleast one uppercase is required")
        elif not has_lowercase:
            print("atleast one lowercase is required")
        elif not has_digits:
            print("atleast one digits is required")
else :
    print("the password must be atleast 8 character".title())
print(passwords)
```
