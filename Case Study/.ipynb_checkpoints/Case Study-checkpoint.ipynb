```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests.exceptions
import sqlite3
import streamlit as st
import seaborn as sns
```

#### Hiting API & Getting Data in json and Converted to csv

```python
base_url = "https://restcountries.com/v3.1/all"
params = {
    "fields": "name,population,region,area,unMember"
}

try:
    print(f"Trying to fetch data with fields parameter...")
    # Pass the parameters using the 'params' argument
    response = requests.get(base_url, params=params) 
    response.raise_for_status() 

    raw_data = response.json()
    print(f"Success! Fetched {len(raw_data)} country records.")

    extracted_data = []
    for country in raw_data:
        extracted_data.append({
            'name': country.get('name', {}).get('common'),
            'region': country.get('region'),
            'population': country.get('population'),
            'area_sqkm': country.get('area'),
            'is_un_member': country.get('unMember')
        })

    df_raw = pd.DataFrame(extracted_data)
    df_raw.to_csv('country_raw_data.csv', index=False)
    print("Data saved to country_raw_data.csv")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

#### Clean csv

```python
df = pd.read_csv('./country_raw_data.csv')

df['population'] = pd.to_numeric(df['population'], errors='coerce')
df['area_sqkm'] = pd.to_numeric(df['area_sqkm'], errors='coerce')

df.dropna(subset=['population', 'area_sqkm'], inplace=True)
print(f"Data shape after cleaning NaNs: {df.shape}")

df['log_population'] = np.log1p(df['population'])
df['log_area_sqkm'] = np.log1p(df['area_sqkm'])

print("Data cleaning and feature engineering complete.")
print(df.head())
```

```python
clean_csv = 'country_clean_data.csv'

df.to_csv(clean_csv, index=False)

print('clean data saved to : {clean_csv}')
```

#### Making plots 

```python
import seaborn as sns

sns.set_style("whitegrid")

df['density'] = df['population'] / df['area_sqkm']

df['log_population'] = np.log1p(df['population'])
df['log_area_sqkm'] = np.log1p(df['area_sqkm'])
```

```python
plt.figure()
plt.figure(figsize=(10, 6))
sns.histplot(df['log_population'], bins=20, kde=True, color='teal')

plt.title('1. Distribution of Log-Transformed Population')
plt.xlabel('Log(Population)')
plt.ylabel('Count of Countries')

plt.show()
```

```python
regional_area = df.groupby('region')['area_sqkm'].mean().sort_values(ascending=False)
plt.figure()
regional_area.plot(kind='bar', color='lightcoral')
plt.title('2. Average Country Area by Region')
plt.ylabel('Average Area (sq km)')
plt.xlabel('Region')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure()
plt.figure(figsize=(8, 6))

plt.boxplot(df['density'].dropna())
plt.title('3. Population Density Distribution')

plt.show()
```

```python
plt.figure()
plt.figure(figsize=(10, 6))

plt.scatter(df['area_sqkm'], df['population'], alpha=0.6)
plt.title('4. Area vs. Population (Scatter Plot)')
plt.xlabel('Area (sq km)')
plt.ylabel('Population')

plt.show()
```

```python
un_pop_sum = df.groupby('is_un_member')['population'].sum()

plt.figure()
plt.figure(figsize=(7, 7))

un_pop_sum.plot(kind='pie', autopct='%1.1f%%', labels=['Non-Member', 'UN Member'], startangle=90)

plt.title('5. Total Population Split by UN Membership')

plt.show()
```

#### Connect to DataBase

```python
clean_csv = pd.read_csv('./country_clean_data.csv')

conn = sqlite3.connect("Country_Data.db")

clean_csv.to_sql('country_table', conn, if_exists='replace', index=False)

sql_query = f"SELECT * FROM {'country_table'}"
df_from_sql = pd.read_sql_query(sql_query, conn)

conn.close()
```
