```python
m = int(input("Enter the numbers of Rows (m) :"))
n = int(input("Enter the numbers of Column (n) :"))
matrix = []

print("Enter the elements row by row (space-separated):")
for i in range(m):
    row = list(map(int, input(f"Enter {n} numbers for Row {i+1}: ").split()))
    if len(row) != n:
        print(f"Error : You must be Enter {n} values.")
        exit()
    matrix.append(row)

for i, row in enumerate(matrix, start=1):
    row_sum = sum(row)
    print(f"Sum of row {i} = {row_sum}")
```

```python
n = int(input("Enter number of rows (n): "))
m = int(input("Enter number of columns (m): "))

def input_matrix(name):
    matrix = []
    print(f"Enter elements of matrix {name} row-wise (space separated):")
    for i in range(n):
        while True:
            row = input(f"Row {i+1}: ").strip().split()
            if len(row) != m:
                print(f"Error: Each row must have exactly {m} elements.")
            else:
                matrix.append([int(x) for x in row])
                break
    return matrix

A = input_matrix('A')
B = input_matrix('B')

result = []
for i in range(n):
    row_sum = []
    for j in range(m):
        row_sum.append(A[i][j] + B[i][j])
    result.append(row_sum)

print("Sum of matrices:")
for row in result:
    print(*row)
```

```python
n = int(input("Enter number of rows (n): "))
m = int(input("Enter number of columns (m): "))

def input_matrix(name):
    matrix = []
    print(f"Enter elements of matrix {name} row-wise (space separated):")
    for i in range(n):
        while True:
            row = input(f"Row {i+1}: ").strip().split()
            if len(row) != m:
                print(f"Error: Each row must have exactly {m} elements.")
            else:
                matrix.append([int(x) for x in row])
                break
    return matrix

A = input_matrix('A')
B = input_matrix('B')

result = []
for i in range(n):
    row_sum = []
    for j in range(m):
        row_sum.append(A[i][j] * B[i][j])
    result.append(row_sum)

print("Multiple of matrices:")
for row in result:
    print(*row)
```
